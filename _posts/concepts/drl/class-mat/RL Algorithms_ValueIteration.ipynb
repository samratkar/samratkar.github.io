Given the specific transition probabilities and rewards, such as the $+2$ reward for going **Fast** while **Cool** and the $-10$ penalty for overheating while **Warm**.

```
import numpy as np
import random
import matplotlib.pyplot as plt

# 1. MDP Definition from Slide
states = ['Cool', 'Warm', 'Overheated']
actions = ['Slow', 'Fast']
gamma = 0.9

transitions = {
    'Cool': {
        'Slow': [(1.0, 'Cool', 1)],
        'Fast': [(0.5, 'Cool', 2), (0.5, 'Warm', 2)]
    },
    'Warm': {
        'Slow': [(0.5, 'Cool', 1), (0.5, 'Warm', 1)],
        'Fast': [(1.0, 'Overheated', -10)]
    },
    'Overheated': {
        'Slow': [(1.0, 'Overheated', 0)],
        'Fast': [(1.0, 'Overheated', 0)]
    }
}

def get_next_state(s, a):
    probs = [t[0] for t in transitions[s][a]]
    idx = np.random.choice(len(probs), p=probs)
    return transitions[s][a][idx][1], transitions[s][a][idx][2]

# --- 2. Algorithms ---

def run_value_iteration(k=20):
    V = {s: 0.0 for s in states}
    history = []
    for _ in range(k):
        new_V = {}
        for s in states:
            if s == 'Overheated':
                new_V[s] = 0
                continue
            new_V[s] = max([sum(p * (r + gamma * V[ns]) for p, ns, r in transitions[s][a]) for a in actions])
        V = new_V
        history.append(V['Cool'])
    return history

def run_q_learning(episodes=500):
    Q = {s: {a: 0.0 for a in actions} for s in states}
    history = []
    alpha, epsilon = 0.1, 0.1
    for _ in range(episodes):
        s, total_r = 'Cool', 0
        while s != 'Overheated':
            a = max(Q[s], key=Q[s].get) if random.random() > epsilon else random.choice(actions)
            ns, r = get_next_state(s, a)
            Q[s][a] += alpha * (r + gamma * max(Q[ns].values()) - Q[s][a])
            s = ns
        history.append(max(Q['Cool'].values()))
    return history

def run_mc_off_policy(episodes=500):
    Q = {s: {a: 0.0 for a in actions} for s in states}
    C = {s: {a: 0.0 for a in actions} for s in states}
    history = []
    for _ in range(episodes):
        episode, s = [], 'Cool'
        for _ in range(20): # Step limit
            if s == 'Overheated': break
            a = random.choice(actions) # Random behavior policy
            ns, r = get_next_state(s, a)
            episode.append((s, a, r))
            s = ns
        G, W = 0.0, 1.0
        for s, a, r in reversed(episode):
            G = r + gamma * G
            C[s][a] += W
            Q[s][a] += (W / C[s][a]) * (G - Q[s][a])
            if a != max(Q[s], key=Q[s].get): break
            W *= 2.0 # Correcting for 0.5 random prob
        history.append(max(Q['Cool'].values()))
    return history

# --- 3. Execution and Visualization ---
vi_hist = run_value_iteration()
td_hist = run_q_learning()
mc_hist = run_mc_off_policy()

plt.figure(figsize=(10, 6))
plt.plot(vi_hist, label='Value Iteration (Exact Math)', linewidth=3, color='black')
plt.plot(td_hist, label='TD Learning (Q-Learning)', alpha=0.7)
plt.plot(mc_hist, label='MC Off-Policy', alpha=0.7)
plt.title('Comparison of RL Methods: Value of "Cool" State')
plt.xlabel('Iterations / Episodes')
plt.ylabel('Estimated Value')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Understanding the Concepts

* **Value Iteration (DP):** This acts as the "ground truth." Since it uses the exact probabilities from your slide, it converges smoothly and quickly to the optimal value.
* **Monte Carlo (MC) Off-Policy:** This uses **Importance Sampling**. It learns from a "behavior policy" (random driving) to estimate the "target policy" (optimal driving). It is often noisier because it relies on full episode completions to update.
* **Temporal Difference (TD/Q-Learning):** This bridges the gap. Like MC, it learns from experience without needing the transition table, but like DP, it "bootstraps" by updating estimates based on other estimates rather than waiting for the episode to end.

### Monto Carlo Off Policy and On Policy Algorithm
We will simulate the Monto Carlo Policy algorithms to see how convergence between behaviour and target policy is achieved using Weighted Importance Sampling(Ordinary, Normalized).

```
import numpy as np
import random

# 1. Setup Environment
states = ['Cool', 'Warm', 'Overheated']
actions = ['Slow', 'Fast']
gamma = 0.9

# Transition Probabilities: P[state][action] = [(prob, next_state, reward), ...]
transitions = {
    'Cool': {
        'Slow': [(1.0, 'Cool', 1)],
        'Fast': [(0.5, 'Cool', 2), (0.5, 'Warm', 2)]
    },
    'Warm': {
        'Slow': [(0.5, 'Cool', 1), (0.5, 'Warm', 1)],
        'Fast': [(1.0, 'Overheated', -10)]
    },
    'Overheated': {
        'Slow': [(1.0, 'Overheated', 0)],
        'Fast': [(1.0, 'Overheated', 0)]
    }
}

def get_next_state(s, a):
    probs = [t[0] for t in transitions[s][a]]
    idx = np.random.choice(len(probs), p=probs)
    return transitions[s][a][idx][1], transitions[s][a][idx][2]

# --- 2. Monte Carlo On-Policy (Epsilon-Greedy) ---
def mc_on_policy(episodes=1000, epsilon=0.1):
    Q = {s: {a: 0.0 for a in actions} for s in states}
    returns = {s: {a: [] for a in actions} for s in states}

    for _ in range(episodes):
        episode = []
        curr_s = 'Cool'
        # Generate an episode
        for _ in range(20): # Limit steps to prevent infinite loops
            if curr_s == 'Overheated': break
            # Epsilon-greedy selection
            a = max(Q[curr_s], key=Q[curr_s].get) if random.random() > epsilon else random.choice(actions)
            next_s, r = get_next_state(curr_s, a)
            episode.append((curr_s, a, r))
            curr_s = next_s

        # Backpropagate returns
        G = 0
        for s, a, r in reversed(episode):
            G = r + gamma * G
            returns[s][a].append(G)
            Q[s][a] = np.mean(returns[s][a])
    return Q

# --- 3. Monte Carlo Off-Policy (Importance Sampling) ---
def mc_off_policy(episodes=1000):
    Q = {s: {a: 0.0 for a in actions} for s in states}
    C = {s: {a: 0.0 for a in actions} for s in states} # Cumulative weights
    target_policy = {s: random.choice(actions) for s in states}

    for _ in range(episodes):
        episode = []
        curr_s = 'Cool'
        # Behavior policy is purely random (b(a|s) = 0.5)
        for _ in range(20):
            if curr_s == 'Overheated': break
            a = random.choice(actions)
            next_s, r = get_next_state(curr_s, a)
            episode.append((curr_s, a, r))
            curr_s = next_s

        G = 0.0
        W = 1.0 # Weight
        for s, a, r in reversed(episode):
            G = r + gamma * G
            C[s][a] += W
            Q[s][a] += (W / C[s][a]) * (G - Q[s][a])
            target_policy[s] = max(Q[s], key=Q[s].get)
            if a != target_policy[s]: break
            W = W * (1.0 / 0.5) # Weight update (1 / prob_behavior)
    return Q

# --- Comparison Execution ---
print("Running Simulations...")
q_on = mc_on_policy()
q_off = mc_off_policy()

print("\n[Monte Carlo On-Policy Q-Values]")
for s in ['Cool', 'Warm']:
    formatted_q_values_on = {a: f"{val:.2f}" for a, val in q_on[s].items()}
    print(f"{s}: {formatted_q_values_on}")

print("\n[Monte Carlo Off-Policy Q-Values]")
for s in ['Cool', 'Warm']:
    formatted_q_values_off = {a: f"{val:.2f}" for a, val in q_off[s].items()}
    print(f"{s}: {formatted_q_values_off}")
```

```
import numpy as np

# 1. MDP Definition
states = ['Cool', 'Warm', 'Overheated']
actions = ['Slow', 'Fast']
gamma = 0.9

# Transition Probabilities: P[state][action] = [(prob, next_state, reward), ...]
transitions = {
    'Cool': {
        'Slow': [(1.0, 'Cool', 1)],
        'Fast': [(0.5, 'Cool', 2), (0.5, 'Warm', 2)]
    },
    'Warm': {
        'Slow': [(0.5, 'Cool', 1), (0.5, 'Warm', 1)],
        'Fast': [(1.0, 'Overheated', -10)]
    },
    'Overheated': {
        # Overheated is a terminal state in many versions,
        # but here we'll assume it loops to itself with 0 reward.
        'Slow': [(1.0, 'Overheated', 0)],
        'Fast': [(1.0, 'Overheated', 0)]
    }
}

# --- 2. Value Iteration ---
def value_iteration(k=10):
    V = {s: 0 for s in states}
    print("Value Iteration Results:")
    for i in range(k):
        new_V = {}
        for s in states:
            if s == 'Overheated':
                new_V[s] = 0
                continue

            action_values = []
            for a in actions:
                ev = sum(prob * (reward + gamma * V[next_s])
                         for prob, next_s, reward in transitions[s][a])
                action_values.append(ev)
            new_V[s] = max(action_values)
        V = new_V
        if i < 3 or i == k-1:
            formatted_V = {s: f"{val:.2f}" for s, val in V.items()}
            print(f"k={i+1}: {formatted_V}")
    return V

# --- 3. Q-Learning (Temporal Difference) ---
def q_learning(episodes=500):
    Q = {s: {a: 0.0 for a in actions} for s in states}
    alpha = 0.1  # Learning rate
    epsilon = 0.1 # Exploration

    for _ in range(episodes):
        curr_s = 'Cool'
        while curr_s != 'Overheated':
            # Epsilon-greedy action selection
            if np.random.rand() < epsilon:
                a = np.random.choice(actions)
            else:
                a = max(Q[curr_s], key=Q[curr_s].get)

            # Simulate step
            probs = [t[0] for t in transitions[curr_s][a]]
            idx = np.random.choice(len(probs), p=probs)
            prob, next_s, reward = transitions[curr_s][a][idx]

            # TD Update: Q(s,a) = Q(s,a) + alpha * [R + gamma * max(Q(s',a')) - Q(s,a)]
            best_next_q = max(Q[next_s].values())
            td_target = reward + gamma * best_next_q
            Q[curr_s][a] += alpha * (td_target - Q[curr_s][a])

            curr_s = next_s
    return Q

# Run Simulations
v_results = value_iteration(k=20)
print("\n---")
q_results = q_learning()
print("Q-Learning (TD) Final Table:")
for s in q_results:
    formatted_q_values = {a: f"{val:.2f}" for a, val in q_results[s].items()}
    print(f"{s}: {formatted_q_values}")
```
