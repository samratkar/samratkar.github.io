```python
import numpy as np
import matplotlib.pyplot as plt

def plot_polynomial(coefficients, x_range=(-10, 10), num_points=1000, title=None):
    """
    Plot a polynomial function.
    
    Parameters:
    -----------
    coefficients : list or array-like
        Coefficients of the polynomial, ordered from highest degree to lowest.
        For example, [3, 2, 1] represents 3x^2 + 2x + 1
    x_range : tuple, optional
        Range of x values to plot (min_x, max_x). Default is (-10, 10).
    num_points : int, optional
        Number of points to use for plotting. Default is 1000.
    title : str, optional
        Title for the plot. If None, a title will be generated based on the polynomial.
    """
    # Convert coefficients to numpy array if not already
    coefficients = np.array(coefficients)
    
    # Generate x values within the specified range
    x = np.linspace(x_range[0], x_range[1], num_points)
    
    # Calculate polynomial values
    y = np.polyval(coefficients, x)
    
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b-', linewidth=2)
    plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
    plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
    plt.grid(True, alpha=0.3)
    
    # Create a polynomial string representation for the title
    if title is None:
        poly_str = ""
        degree = len(coefficients) - 1
        
        for i, coef in enumerate(coefficients):
            if coef == 0:
                continue
                
            term_degree = degree - i
            
            if coef < 0:
                sign = " - " if i > 0 else "-"
                coef = abs(coef)
            else:
                sign = " + " if i > 0 else ""
            
            if term_degree > 1:
                poly_str += f"{sign}{coef}x^{term_degree}"
            elif term_degree == 1:
                poly_str += f"{sign}{coef}x"
            else:
                poly_str += f"{sign}{coef}"
        
        title = f"Plot of polynomial: {poly_str}"
    
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('y')
    
    # Set reasonable axis limits based on the data
    y_range = max(abs(np.max(y)), abs(np.min(y)))
    y_limit = min(100, y_range * 1.1)  # Limit the view to prevent extreme scales
    plt.ylim(-y_limit, y_limit)
    
    plt.show()

# Example usage:
if __name__ == "__main__":
    
    # Example 1: Plot x^3 - 6x^2 + 11x - 6
    print("Example 1: Plotting x^3 - 6x^2 + 11x - 6")
    plot_polynomial([1, -6, 11, -6])
    
    # Example 2: Plot 2x^4 - 3x^2 + 1
    print("Example 2: Plotting 2x^4 - 3x^2 + 1")
    plot_polynomial([2, 0, -3, 0, 1])

    # Example 3: Plot 2x - 3
    print("Example 2: Plotting 2x^4 - 3x^2 + 1")
    plot_polynomial([2, -3])
```
